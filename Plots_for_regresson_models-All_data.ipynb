## Cumulative Visualizations for Sigma, R, and RG parameters

```python
all_models = []

all_models.append(get_true_predicted_crossvalid('diffuse', df))
all_models.append(get_true_predicted_crossvalid('diffuse__diffuse', df))
all_models.append(get_true_predicted_crossvalid('crystalline', df))
all_models.append(get_true_predicted_crossvalid('disordered', df))
all_models.append(get_true_predicted_crossvalid('diffuse__disordered', df))
all_models.append(get_true_predicted_crossvalid('diffuse__crystalline', df))
all_models.append(get_true_predicted_crossvalid('diffuse__diffuse__crystalline', df))
all_models.append(get_true_predicted_crossvalid('diffuse__diffuse__disordered', df))
all_models.append(get_true_predicted_crossvalid('diffuse__disordered__crystalline', df))
```

```python
sigma_models = {}
r_models = {}
rg_models = {}
for gr in all_models:
    for k,v in gr.items():
        if k.endswith('_sigma'):
            sigma_models[k] = v
        elif k.endswith('_rg'):
            rg_models[k] = v
        elif k.endswith('_r'):
            r_models[k] = v
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,10))
c = 0
for k, v in sigma_models.items():
    x = v['true_y']
    y = v['cross_val']
    z = v['predicted']
    if c == 0:
        plt.scatter(x, y, marker='^', c='green', label =  'crossvalidated')
        plt.scatter(x, z, marker='o',c='blue', label = 'predicted')
        c = 1
    else:
        plt.scatter(x, y, marker='^', c='green')
        plt.scatter(x, z, marker='o',c='blue')
    plt.plot(x, x, 'b--')
    ax.legend()
ax.set_title('Predicted vs Crossvalidated values for R_Normal_Sigma', fontsize=20)
fig.savefig('pr_vs_crossval_sigma.pdf')
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,10))
c= 0
for k, v in r_models.items():
    x = v['true_y']
    y = v['cross_val']
    z = v['predicted']
    if c == 0:
        plt.scatter(x, y, marker='^', c='green', label =  'crossvalidated')
        plt.scatter(x, z, marker='o',c='blue', label = 'predicted')
        c = 1
    else:
        plt.scatter(x, y, marker='^', c='green')
        plt.scatter(x, z, marker='o',c='blue')
    plt.plot(x, x, 'b--')
    ax.legend()
ax.set_title('Predicted vs Crossvalidated values for R', fontsize=20)
fig.savefig('pr_vs_crossval_r.pdf')
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,10))
c= 0
for k, v in rg_models.items():
    x = v['true_y']
    y = v['cross_val']
    z = v['predicted']
    if c == 0:
        plt.scatter(x, y, marker='^', c='green', label =  'crossvalidated')
        plt.scatter(x, z, marker='o',c='blue', label = 'predicted')
        c = 1
    else:
        plt.scatter(x, y, marker='^', c='green')
        plt.scatter(x, z, marker='o',c='blue')
    plt.plot(x, x, 'b--')
    ax.legend()
ax.set_title('Predicted vs Crossvalidated values for RG', fontsize=20)
fig.savefig('pr_vs_crossval_rg.pdf')
```
